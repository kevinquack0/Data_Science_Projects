
# Matplotlib 


```python
import numpy as np
x = np.arange(0,100)
y = x*2
z = x**2
```

Imports

```python
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

Basics

```python
fig = plt.figure()
ax= fig.add_axes([0,0,1,1])
ax.plot(x,y)
ax.set_xlabel("x")
ax.set_ylabel("y")
ax.set_title("title")
```


Figure object and put two axes on it, ax1 and ax2. Located at [0,0,1,1] and [0.2,0.5,.2,.2] respectively

```python
fig = plt.figure()
ax1= fig.add_axes([0,0,1,1])
ax2 = fig.add_axes([0.2,0.5,0.2,0.2])
```

Ploting (x,y) on both axes.

```python
fig = plt.figure()
ax1= fig.add_axes([0,0,1,1])
ax2 = fig.add_axes([0.2,0.5,0.2,0.2])
ax1.plot(x,y)
ax2.plot(x,y)
```

Ploting below by adding two axes to a figure object at [0,0,1,1] and [0.2,0.5,.4,.4]

```python
fig = plt.figure()
ax1= fig.add_axes([0,0,1,1])
ax2 = fig.add_axes([0.2,0.5,0.4,0.4])
```

Using x,y, and z arrays to create a plot

```python
fig = plt.figure()
ax1= fig.add_axes([0,0,1,1])
ax2 = fig.add_axes([0.2,0.5,0.4,0.4])
ax1.plot(x,z)
ax2.set_xlim([20,22])
ax2.set_ylim([30,50])
ax2.plot(x,y)
ax2.set_title("zoom")
ax2.set_xlabel("x")
ax2.set_ylabel("y")
ax1.set_xlabel("x")
ax1.set_ylabel("z")

```

Create the subplots 

 fig, axes = plt.subplots(1, 3, figsize=(4, 4))

```python
fig, axes = plt.subplots(1, 2)
```

Ploting (x,y) and (x,z) with styles.

```python
fig, axes = plt.subplots(1, 2)
axes[0].plot(x,y,ls="--",lw=4)

axes[1].plot(x,z,color="red",lw=4)
```

Using resize

```python
fig, axes = plt.subplots(1, 2,figsize=(12,2))
axes[0].plot(x,y,ls="--",lw=4)
axes[1].plot(x,z,color="red",lw=4)
```
